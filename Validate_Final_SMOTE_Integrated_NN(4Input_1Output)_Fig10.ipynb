```python
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
# Load the Keras HDF5 model
model = keras.models.load_model('my_model7.h5')
### 1st plot
Temperature = 324.37
Gas_flow_rate = 50
Ratio_exp = 1
Pressure_range = range(0, 81, 1)
Combined_array = []

for Pressure in Pressure_range:
      Combined_array.append([Temperature, Pressure, Ratio_exp, Gas_flow_rate])

z_pred = model.predict(Combined_array)

# Extract pressure from Combined_array
pressure = [row[1] for row in Combined_array]

# Create 2D figure
plt.figure(3)
figure = plt.plot(pressure, z_pred, label='R=1, T = 324.37 C, Flow rate = 3000ml/h.gcat')
plt.xlabel('Pressure(psig)')
plt.ylabel('NH3 concentration(%)')
plt.legend()

### 2nd plot
Pressure = 80
Gas_flow_rate = 50
Ratio_exp = 1
Temperature_range = range(280, 401, 1)
Combined_array = []

for Temperature in Temperature_range:
      Combined_array.append([Temperature, Pressure, Ratio_exp, Gas_flow_rate])

z_pred = model.predict(Combined_array)

# Extract temperature from Combined_array
temperature = [row[0] for row in Combined_array]

# Create 2D figure
plt.figure(4)
figure = plt.plot(temperature, z_pred, label='R=1, P = 80psig, Flow rate = 3000ml/h.gcat')
plt.xlabel('Temperature(C)')
plt.ylabel('NH3 concentration(%)')
plt.legend()

### 3rd plot
Pressure = 80
Gas_flow_rate = 50
Temperature = 324.37
num_elements = int((5 - 0.50) / 0.1) + 1
Ratio_exp = np.linspace(0.50, 5, num_elements)
Combined_array = []

for Ratio in Ratio_exp:
      Combined_array.append([Temperature, Pressure, Ratio, Gas_flow_rate])

z_pred = model.predict(Combined_array)

# Extract ratio from Combined_array
ratio = [row[2] for row in Combined_array]

# Create 2D figure
plt.figure(5)
figure = plt.plot(ratio, z_pred, label='Temperature=324.37 C, P = 80psig, Flow rate = 3000ml/h.gcat')
plt.xlabel('H2 to N2 ratio')
plt.ylabel('NH3 concentration(%)')
plt.legend()

### 4th plot
Temperature = 324.3697479
Pressure = 80
Ratio_exp = 1
Gas_flow_range = range(50, 250, 10)
Combined_array = []

for Gas_flow in Gas_flow_range:
      Combined_array.append([Temperature, Pressure, Ratio_exp, Gas_flow])

z_pred = model.predict(Combined_array)

# Extract pressure from Combined_array
gas_flow = [row[3] for row in Combined_array]

# Create 2D figure
plt.figure(6)
figure = plt.plot(gas_flow, z_pred, label='R=1, T = 324.37 C, Pressure = 80psig')
plt.xlabel('Gas flor rate(ml/h.gcat)')
plt.ylabel('NH3 concentration(%)')
plt.legend()
```
